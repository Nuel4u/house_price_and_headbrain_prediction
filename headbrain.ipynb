```python
x =df['Head Size(cm^3)'].values
y =df['Brain Weight(grams)'].values
```

```python
mean_x =np.mean(x)
mean_y =np.mean(y)
```

```python
mean_x
```

```python
mean_y
```

```python
m =len(x)  
```

```python
number = 0
denum = 0

for i in range(m):
    number +=(x[i] - mean_x) * (y[i] - mean_y)      #(x-x`)(y-y`)
    denum +=(x[i] - mean_x) ** 2                    #(x-x`)pow2
```

```python
b1 = number /denum             # m= number / denum
b0 = mean_y - (b1 * mean_x)     # y  = m *x +c =  (c = y -m*x)
```

```python
b1   # value of m     - head size of linear model
```

```python
b0 # value of c      - brainweight of linear model
```

```python
plt.plot(x,y ,color = '#58b970' , label ='Regression Line')  # ploting lines
plt.scatter(x,y,c='#ef5423' , label ='scatter plot')


plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams')
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred += b0 +b1 * x[i]
    ss_t +=(y[i] - mean_y) **2
    ss_r +=(y[i] - y_pred) ** 2
    
r2 =1 -(ss_r/ss_t)
print(r2)
```
