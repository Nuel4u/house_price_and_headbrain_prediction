```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df =pd.read_excel('Documents//datasets//housebook.xlsx')
```

```python
df.head()
```

# getties dummy values

```python
dummies=pd.get_dummies(df.town)
```

```python
dummies
```

# concatinating dummies with df values

```python
merged = pd.concat([df ,dummies])
```

```python
merged
```

```python
merged = pd.concat([df ,dummies] ,axis='columns')
```

```python
merged
```

# dropping columns

```python
final=merged.drop(['town' ,'westwindsor'],axis="columns")
```

```python
final
```

# developing linear model

```python
reg = linear_model.LinearRegression()
```

# choosing and dropping  data for modeling

```python
x = final.drop('price' ,axis="columns")
```

```python
x
```

# dependent variable 

```python
y =final.price
```

```python
y
```

# training and fitting data

```python
reg.fit(x,y)
```

# predicting prices

```python
reg.predict([[2800 ,1 ,0,1,0,0]]) # each parameter represent row of each table except price
```

```python
reg.predict([[3700 ,0 ,1,0,1,0]])
```

# acuracy of the model

```python
reg.score(x,y)*100
```
